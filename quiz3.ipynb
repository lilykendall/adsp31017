```python
import pandas as pd

df = pd.DataFrame({
    'Target Category': ['I', 'II', 'III', 'IV', 'V'],
    'Count': [6606, 11324, 7280, 2080, 223]
})

# train MNLogit model intercept only
# reference label is category II, 
# what is the intercept of category V?
import statsmodels.api as sm
model = sm.MNLogit(df['Count'], sm.add_constant(df['Target Category'].map({'I': 0, 'II': 1, 'III': 2, 'IV': 3, 'V': 4})))
result = model.fit()
intercept_category_V = result.params.loc['const', 'Target Category[5]']
print(intercept_category_V)
```
